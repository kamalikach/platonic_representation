```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_file(fname):
    x, y = np.loadtxt(fname, delimiter=",", unpack=True)

    plt.scatter(x, y)
    plt.show()

    r = np.corrcoef(x, y)[0, 1]
    print("R", r)
```

```python
plot_file("./outputs/cifar_resnet_alexnet_35_2026-02-19_20-59-33.txt")
```
